```python
import pandas as pd   
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.stattools import adfuller
import statsmodels.api as sm
from scipy import stats
```

#Without Diff

```python
df_trends = pd.read_excel('/content/World_GT_Full.xlsx')
```

```python
df_trends['Date'] = pd.to_datetime(df_trends.iloc[:, 0], infer_datetime_format=True)
```

```python
df_trends = df_trends.set_index(df_trends.iloc[:,0])
df_trends = df_trends.drop(['Date'], axis = 1)
```

```python
df_trends
```

```python
stati = []
trend_names = list(df_trends.columns)
adf = []
p_val = []
critical_value = []
```

```python
#Storing Results in Data Frame

for column in df_trends:
  result = adfuller(df_trends[column])
  stati.append(result)
  adf.append(result[0])
  p_val.append(result[1])
  critical_value.append(result[4])
```

```python
stationary_tests = pd.DataFrame(columns=['Trend Name','ADF Statistic','p-value','Critical Values'])
stationary_tests['ADF Statistic'] = adf
stationary_tests['p-value'] = p_val
stationary_tests['Critical Values'] = critical_value
stationary_tests['Trend Name'] = trend_names
stationary_tests
```

```python
stationary_tests = stationary_tests.drop(['Critical Values'], axis=1)
```

```python
stationary_tests
```

```python
#Exporting to LaTEX

with open('mytable2.tex', 'w') as tf:
     tf.write(stationary_tests.to_latex(caption='This is my caption', label='Table2', column_format='c|c|c|c'))
```

#With Diff()

```python
df_trends = pd.read_excel('/content/World_GT_Full.xlsx')
```

```python
df_trends['Date'] = pd.to_datetime(df_trends.iloc[:, 0], infer_datetime_format=True)
df_trends = df_trends.set_index(df_trends.iloc[:,0])
df_trends = df_trends.drop(['Date'], axis = 1)
```

```python
df_trends.iloc[:,1:] = df_trends.iloc[:,1:].diff()
df_trends = df_trends.dropna()
df_trends
```

```python
stati = []
trend_names = list(df_trends.columns)
adf = []
p_val = []
critical_value = []
```

```python
#Storing Results in Data Frame

for column in df_trends:
  result = adfuller(df_trends[column])
  stati.append(result)
  adf.append(result[0])
  p_val.append(result[1])
  critical_value.append(result[4])
```

```python
stationary_tests_diff = pd.DataFrame(columns=['Trend Name','ADF Statistic','p-value','Critical Values'])
stationary_tests_diff['ADF Statistic'] = adf
stationary_tests_diff['p-value'] = p_val
stationary_tests_diff['Critical Values'] = critical_value
stationary_tests_diff['Trend Name'] = trend_names
stationary_tests_diff
```

```python
stationary_tests_diff = stationary_tests_diff.drop(['Critical Values'], axis=1)
stationary_tests_diff
```

```python
#Exporting to LaTEX

with open('mytable1.tex', 'w') as tf:
     tf.write(stationary_tests_diff.to_latex(caption='This is my caption', label='Table2', column_format='c|c|c|c'))
```
