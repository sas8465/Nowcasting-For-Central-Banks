#Stationary Tests

```python
from statsmodels.tsa.stattools import adfuller
```

```python
ADF_USD = adfuller(include_chfusd['Adj Close'])
ADF_EUR = adfuller(include_eurchf['CHF/EUR'])
ADF_USDLog = adfuller(include_chfusd['Log_Difference'])
ADF_EURLog = adfuller(include_eurchf['log_difference'])
ADF_DBV = adfuller(df_dbv['Adj Close'])
ADF_DBVLog = adfuller(df_dbv['Log Difference'])
```

```python
print('critical value: ', ADF_USD[0])
print('p-value: ', ADF_USD[1])
print('5% Significance ADF Statistic: ', ADF_USD[4]['5%'])
```

```python
exchange_rate_stationary_test = { 'Exchange Rates': ['CHF/USD', 'CHF/EUR', 'DBV Port.'],
                                  'ADF Statistic': [ADF_USD[0],ADF_EUR[0],ADF_DBV[0]],
                                  'p-value': [ADF_USD[1], ADF_EUR[1], ADF_DBV[1]],
                                  '5% Significance ADF Critical Value': [ADF_USD[4]['5%'], ADF_EUR[4]['5%'], ADF_DBV[4]['5%']],
                                  'ADF Statistic 1': [ADF_USDLog[0],ADF_EURLog[0], ADF_DBVLog[0]],
                                  'p-value 1': [ADF_USDLog[1], ADF_EURLog[1], ADF_DBVLog[1]],
                                  '5% Significance ADF Critical Value 1': [ADF_USDLog[4]['5%'], ADF_EURLog[4]['5%'], ADF_DBVLog[4]['5%']]
                                  }

exchange_rate_stationary_table = pd.DataFrame(exchange_rate_stationary_test)
```
